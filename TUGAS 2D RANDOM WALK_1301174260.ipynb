```python
# Python code for 2D random walk.
import numpy
import pylab
import matplotlib.pyplot as plt
from random import randint
import math
import random

#pertama import library
```

```python
#menentukan arah pergerakan dari partikel sebanyak 8 arah
def update(list):
    rand = random.uniform(0,1)
    rand = round(rand,2)
    if (rand <=0.125):
        list[1] = list[1]+1
        list[0] = list[0]+1 #kananatas
    elif (rand <=0.25):
        list[0] = list[0]+1 #kanan
    elif (rand <=0.375):
        list[0] = list[0]+1 
        list[1] = list[1]-1 #kananbawah
    elif (rand <=0.5):
        list[1] = list[1]-1 #bawah
    elif (rand <=0.6255):
        list[0] = list[0]-1  
        list[1] = list[1]-1 #kiribawah
    elif (rand <=0.75):
        list[0] = list[0]-1 #kiri  
    elif (rand <=0.875):
        list[1] = list[1]+1  
        list[0] = list[0]-1 #kiriatas
    else:
        list[1] = list[1]+1 #atas
```

```python
#memberikan batasan agar tiap partikel tetap berada pada skala x dan y
def pbc(list):
    # x axis
    if list[0] > x_max:
        list[0] = list[0] - x_range
    if list[0] < x_min:
        list[0] = list[0] + x_range
    # y axis
    if list[1] > y_max:
        list[1] = list[1] - y_range
    if list[1] < y_min:
        list[1] = list[1] + y_range    
    
```

```python
#memberikan warna kepada tiap 10 partikel
def coloring(col):
    col = {0: 'maroon', 1: 'forestgreen', 2: 'darkslategray', 3: 'indigo', 4: 'k', 5: 'sienna', 6: 'darkgoldenrod', 7: 'darkcyan', 8: 'crimson', 9: 'y'}
    return col
```

```python
#menentukan ukuran dari x dan y
x_min = 0
x_max = 30
y_min = 0
y_max = 30
x_range = x_max - x_min
y_range = y_max - y_min
```

```python
#menentukan jumlah partikel
n_particle=10
pos = []
pos_awal = []
for i in range (0,n_particle):
    rand1 = randint(0,30)
    rand2 = randint(0,30)
    pos.append([rand1,rand2])
    pos_awal.append([rand1,rand2])
```

```python
#menentukan titik mula 10 partikel
col = {}
color = coloring(col)
for i in range(len(pos)):
    plt.scatter(pos[i][0],pos[i][1], color=color[i])
plt.title('Titik mula 10 Partikel')
```

```python
#memberikan hasil berupa output ketika partikel telah melangkah sebanyak 100
n_iterasi = 100
for i in range(n_iterasi):
  for j in range(len(pos)):
      x = pos[j][0]
      y = pos[j][1]
      update(pos[j])
      x_update = pos[j][0]
      y_update = pos[j][1]
      pbc(pos[j])
      x_pbc = pos[j][0]
      y_pbc = pos[j][1]
      hasil_x = [x, x_pbc]
      hasil_y = [y, y_pbc]
      if(x_update==x_pbc and y_update==y_pbc):
          plt.plot(hasil_x, hasil_y, color=color[j])
      if(i==(n_iterasi-1)):
          plt.scatter(pos[j][0],pos[j][1], color=color[j], linestyle = '--')

  plt.title("100 Langkah pergerakan dari 10 Partikel")
  plt.show
```
